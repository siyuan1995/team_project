```python
#Analysis of Factors Affecting Average Credit Limit
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file from the given path
file_path = r"../data/raw/BankChurners.csv"
df = pd.read_csv(file_path)

# Function to create and display bar plots
def create_bar_plot(data, x_var, y_var='Credit_Limit'):
    plt.figure(figsize=(10, 6))
    sns.barplot(data=data, x=x_var, y=y_var, order=data.sort_values(by=y_var)[x_var])
    plt.title(f'Average {y_var} by {x_var}')
    plt.xticks(rotation=45)
    plt.show()

# List of variables to analyze
variables = ['Gender', 'Dependent_count', 'Education_Level', 'Marital_Status', 'Income_Category',
             'Card_Category', 'Months_on_book', 'Total_Relationship_Count', 'Months_Inactive_12_mon', 
             'Contacts_Count_12_mon']

# Calculate the mean Credit_Limit for each variable and create bar plots
for var in variables:
    pivot_table = df.pivot_table(values='Credit_Limit', index=var, aggfunc='mean').reset_index()
    create_bar_plot(pivot_table, var)
```

Based on the provided visualizations, here are the conclusions about average credit limits among various variables:

1.Gender: Males have a significantly higher average credit limit than females.
2.Dependent Count: Average credit limit generally increases with the number of dependents, with higher dependents having a higher credit limit.
3.Education Level: There is a relatively consistent average credit limit across different education levels.
4.Marital Status: Single, divorced, and unknown marital statuses have slightly higher average credit limits compared to married individuals.
5.Income Category: Higher income categories correlate with higher average credit limits, with the highest limits observed in the $120K+ category.
6.Card Category: Higher-tier card categories (Platinum, Gold) have significantly higher average credit limits compared to lower-tier categories (Blue).
7.Months on Book: Average credit limit increases with the number of months on book, indicating that longer-term customers tend to have higher credit limits.
8.Total Relationship Count: More relationships with the bank correlate with higher average credit limits.
9.Months Inactive in 12 Months: Customers with fewer inactive months generally have lower credit limits compared to those with higher inactive months.
10.Contacts Count in 12 Months: More frequent contacts with the bank correlate with higher average credit limits.


Variables used in the Analysis:
Target Variable:           	
Credit_Limit (int64): Customer credit limit in US dollars.
Potential Predictors:
Customer_Age (int64): Customer age in years.
Gender (Categorical): Customer gender defined in binary terms Male and Female (Categories: M, F)
Dependent_count (int64): Customer’s number of dependents.
Months_on_book (int64): Number of months since customer's card first activation
Months_Inactive_12_mon (int64): Number of months without activity in customer’s card in the past 12 months
Avg_Utilization_Ratio (float64): Average proportion of customer’s credit card limit usage within the past
Total_Revolving_Balance (int64): Average revolving balance in customer’s card within the past 12 months
Total_Relationship_Count (int64): Number of open products (accounts, credits, etc.) that the customer has with the bank
Education_Level (Categorical): Customer education level (Categories: Uneducated, High School, College, Graduate, Post-Graduate, Doctorate, Unknown)
Marital_Status (Categorical): Customer Marital Status (Categories: Single, Married, Divorced, Unknown)
Income (Categorical): Customer’s income range in dollars (Categories: Less than $40K, $40K - $60K, $60K - $80K, $80K - $120K, $120K +)
